```python
import pandas as pd
import csv

game_1 = "../resources/pymoli.csv"
game_data = pd.read_csv(game_1)
game_data.head()
```

```python
tot_players = game_data['SN'].unique()

summary_df = pd.DataFrame({"Total Players": [len(tot_players)]})

summary_df
```

```python
unique_item = game_data["Item ID"].unique()
unique_item

avg_price = game_data["Price"].mean()
round(avg_price, 2)

num_purchases = game_data["Purchase ID"].count()
num_purchases

tot_rev = game_data["Price"].sum()
tot_rev

summary2_df = pd.DataFrame({"Number Of Unique Items": [len(unique_item)], "Average Price": [round(avg_price, 2)], "Number Of Purchases": [num_purchases], "Total Revenue": [tot_rev]})
summary2_df
```

```python
gen_ref = game_data["Gender"].value_counts()
gen_sum = game_data["Gender"].count()
pct = gen_ref.div(gen_sum).apply('{:.0%}'.format)

playerdem_df = pd.DataFrame({"Total Count": [gen_ref["Male"], gen_ref["Female"], gen_ref["Other / Non-Disclosed"]], "Percentage Of Players": [pct["Male"], pct["Female"], pct["Other / Non-Disclosed"]]}, index=['Male', 'Female', 'Other / NA'])
playerdem_df
```

```python
df_group = game_data.groupby(['Gender'])
df2_group = game_data.groupby(['SN', 'Gender'])
print(df_group)

tpv = df_group["Price"].sum()
app = df_group["Price"].mean()
pc = df_group["Purchase ID"].count()
new = df2_group["Price"].sum()
newnew = new.groupby(['Gender'])
newnewnew = newnew.mean()
summ_df3 = pd.DataFrame({"Purchase Count": pc, "Avg Purchase Price": round(app, 2), "Total Purchase Value": tpv, "Avg Total Purchase per Person": round(newnewnew, 2)})
summ_df3
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
game_data["Age Ranges"] = pd.cut(game_data["Age"], bins, labels=group_names)
game_test1 = game_data.drop_duplicates(['SN'], keep='last')

ggame_data = game_test1.groupby(["Age Ranges"])
c2game_data = (ggame_data["Age"].count() / game_test1["Age"].count()).apply('{:.0%}'.format)
summ_df4 = pd.DataFrame({"Total Count": cgame_data, "Percentage of Players": c2game_data})
summ_df4
```

```python
summ5 = game_data.groupby(["Age Ranges"])
summ6 = summ5["Purchase ID"].count()
summ7 = round(summ5["Price"].mean(), 2)
summ8 = summ5["Price"].sum()
summ9 = game_data.groupby(['SN', 'Age Ranges'])
summ10 = summ9["Price"].sum()
summ11 = summ10.groupby(['Age Ranges'])
summ12 = summ11.mean()
summ_df5 = pd.DataFrame({"Purchase Count": summ6, "Average Purchase Price": summ7, "Total Purchase Value": summ8, "Avg Total Purchase per Person": round(summ12, 2)})
summ_df5
```
